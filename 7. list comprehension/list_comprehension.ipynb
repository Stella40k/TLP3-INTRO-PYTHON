# List Comprehension en Python

## ¿Qué es?

**List Comprehension** es una forma concisa y elegante de crear listas en Python. Permite construir una nueva lista aplicando una expresión a cada elemento de una secuencia iterable.

## Sintaxis Básica

```python
nueva_lista = [expresión for elemento in iterable]
```

## List Comprehension vs Bucle Tradicional

| Aspecto | Bucle for | List Comprehension |
|---------|-----------|-------------------|
| Líneas de código | Más | Menos |
| Legibilidad | Más clara en casos complejos | Ideal para casos simples |
| Velocidad | Normal | Más rápida |
