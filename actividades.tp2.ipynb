Tienes una lista de productos con su `nombre`, su `stock` y su `precio`.

a) Calcula una lista (mediante comprensión de listas) que incluya el nombre de cada producto, junto con el **valor total** del stock que se tiene de cada uno.

```python
productos = [
    {'nombre': 'Arroz',     'precio': 25,  'stock': 50},
    {'nombre': 'Aceite',    'precio': 80,  'stock': 0},
    {'nombre': 'Pasta',     'precio': 30,  'stock': 20},
    {'nombre': 'Queso',     'precio': 120, 'stock': 8},
    {'nombre': 'Sal',       'precio': 10,  'stock': 100},
    {'nombre': 'Cafe',      'precio': 95,  'stock': 0},
    {'nombre': 'Harina',    'precio': 22,  'stock': 15},
]

valores_totales = [{'nombre': p['nombre'], 'valor': p['precio'] * p['stock']} for p in productos] 

# p['nombre'] le dice donde ir a buscar, tal producto y trae el texto
# p['precio'] * p['stock'] le dice q multiplique los numberos q encuentre en esas etiquetas
# for p in productos es un bucle, p es una variable temporal, en la primera vuelta es el arroz. Hace 
# el calculo y lo guarda, en la segunda pasa con otro producto y asi

print(valores_totales)
```

b) Retorna una lista de los productos con un **stock menor a 10** y con un **precio mayor a 20**. usando comprensión de listas.

```python
productos = [
    
    {'nombre': 'Arroz',     'precio': 25,  'stock': 50},
    {'nombre': 'Aceite',    'precio': 80,  'stock': 0},
    {'nombre': 'Pasta',     'precio': 30,  'stock': 20},
    {'nombre': 'Queso',     'precio': 120, 'stock': 8},
    {'nombre': 'Sal',       'precio': 10,  'stock': 100},
    {'nombre': 'Cafe',      'precio': 95,  'stock': 0},
    {'nombre': 'Harina',    'precio': 22,  'stock': 15},
]

productos_con_bajo_stock = [p for p in productos if p['stock'] < 10 and p['precio'] > 20]
#con solo la condicion y el and pude recorrer la lista y cumplir la condicion sin el uso tradicional del for

print(productos_con_bajo_stock)
```

c) Se quiere realizar una rebaja por liquidación de los productos con menor stock. 
   Calcula una lista (usando comprensión de listas) de los productos que **tienen 20 en stock o menos**.
   Además, **aplica un descuento del 15% si el precio es mayor a 50**. 
   
   **Pista**: Puedes definir una función para aplicar el descuento basándote en el precio, luego utiliza la comprensión de lista para filtrar los productos.

```python
productos = [
    
    {'nombre': 'Arroz',     'precio': 25,  'stock': 50},
    {'nombre': 'Aceite',    'precio': 80,  'stock': 0},
    {'nombre': 'Pasta',     'precio': 30,  'stock': 20},
    {'nombre': 'Queso',     'precio': 120, 'stock': 8},
    {'nombre': 'Sal',       'precio': 10,  'stock': 100},
    {'nombre': 'Cafe',      'precio': 95,  'stock': 0},
    {'nombre': 'Harina',    'precio': 22,  'stock': 15},
]
descuento = 0.85 

#usar funncion para separar la logica de compra del escuento

#def para crear la funcion, el nombre y entre ´parentesis los parametros(la entrada de la fun)
def aplicar_descuento(calcular_descuento):
    if calcular_descuento > 50:
        return calcular_descuento * descuento
    else:
        return calcular_descuento

productos_en_liquidacion = [{'nombre': p['nombre'],
                             'precio_final': aplicar_descuento(p['precio'])
                            }
                            for p in productos 
                            if p['stock'] <= 20
]
#llamo a la funcion, el precio es el parametro. Compara el valor del parametro, si es mayor a 
#50 aplica el descuento, hace el calculo y devuelve un nuevo recio. Sino, pasa directo
print(productos_en_liquidacion)
```

Se tomaron los siguientes datos en sensores de temperatura a lo largo del país. Sin embargo, hay datos faltantes y valores inválidos.

```python
lecturas = [
    {'temperatura': 25,  'humedad': 60},
    {'temperatura': 55,  'humedad': 88},
    {'temperatura': -15, 'humedad': 45},
    {'temperatura': 38,  'humedad': 95},
    {'temperatura': 42,  'humedad': 101},
    {'temperatura': None,  'humedad': 101},
    {'temperatura': 22,  'humedad': 35},
    {'temperatura': 35,  'humedad': None},
    {'temperatura': 61,  'humedad': None},
    {'temperatura': 100,  'humedad': 102},
]
```

a) Filtra y calcula una lista de sólamente las temperaturas **válidas**: aquellas que no son `None` y que se encuentran entre -20 y 60 inclusive. Además, convierte la temperatura de grados Celsius a Fahrenheit, aplicando la fórmula: `F = C * 9/5 + 32`.

```python
lecturas = [
    {'temperatura': 25,  'humedad': 60},
    {'temperatura': 55,  'humedad': 88},
    {'temperatura': -15, 'humedad': 45},
    {'temperatura': 38,  'humedad': 95},
    {'temperatura': 42,  'humedad': 101},
    {'temperatura': None,  'humedad': 101},
    {'temperatura': 22,  'humedad': 35},
    {'temperatura': 35,  'humedad': None},
    {'temperatura': 61,  'humedad': None},
    {'temperatura': 100,  'humedad': 102},
]

#misma logica q las notas pero la limpiamo con una condicion de q no sea none y 
#q este entre klos -20 u 60

#la formula es para conversion matematica de celcius a faren
temperaturas_validas = [lecturas['temperatura'] * 9/5 + 32
                        for lecturas in lecturas if lecturas['temperatura'] 
                        is not None and -20 <= lecturas['temperatura'] <= 60
                        ]
print(temperaturas_validas)
```

b) Filtra las lecturas descartando aquellas que tengan datos faltantes (`None`) o valores inválidos. Se considera inválido:
- Una temperatura fuera del rango -20 a 60 (inclusive)
- Una humedad mayor a 100

Retorna una lista con los diccionarios de las lecturas que superaron el filtro.

```python
lecturas = [
    {'temperatura': 25,  'humedad': 60},
    {'temperatura': 55,  'humedad': 88},
    {'temperatura': -15, 'humedad': 45},
    {'temperatura': 38,  'humedad': 95},
    {'temperatura': 42,  'humedad': 101},
    {'temperatura': None,  'humedad': 101},
    {'temperatura': 22,  'humedad': 35},
    {'temperatura': 35,  'humedad': None},
    {'temperatura': 61,  'humedad': None},
    {'temperatura': 100,  'humedad': 102},
]

#limpiar datos con error

#devuelvo el diccionario completo
#si el elemento pasa los filtros me tendra q dar el objeto completo
#en la linea del if es si la temperatura falta, lo descarta direcctamente
#la liea del and le obliga que todos deben ser validos
lecturas_validas = [lecturas for lecturas in lecturas 
                    if lecturas['temperatura']
                    is not None and -20 <= lecturas['temperatura'] <= 60
                    and lecturas['humedad'] is not None and lecturas['humedad'] <= 100
                    ]
print(lecturas_validas)
```

c) A partir de `lecturas_validas`, calcula el nivel de criticidad de cada lectura:
- Si la temperatura es mayor a 50 **o** la humedad es mayor a 90, el nivel es `'critico'`.
- En cualquier otro caso, el nivel es `'normal'`.

Retorna una lista de strings con el nivel de cada lectura válida.

**Pista**: Puedes definir una función que reciba una lectura y devuelva su nivel, y luego usarla dentro de la comprensión de listas.

**Desafío adicional (opcional)**: En lugar de retornar un string, retorna un diccionario. Una de las claves será `nivel`. Además, incluye las claves `temperatura` o `humedad` cuyos valores hayan superado los umbrales normales. Por ejemplo, para `{'temperatura': 57, 'humedad': 87}` el resultado sería `{'nivel': 'critico', 'temperatura': 57}`. Si ambos valores son anormales, incluye ambas claves.

```python
lecturas = [
    {'temperatura': 25,  'humedad': 60},
    {'temperatura': 55,  'humedad': 88},
    {'temperatura': -15, 'humedad': 45},
    {'temperatura': 38,  'humedad': 95},
    {'temperatura': 42,  'humedad': 101},
    {'temperatura': None,  'humedad': 101},
    {'temperatura': 22,  'humedad': 35},
    {'temperatura': 35,  'humedad': None},
    {'temperatura': 61,  'humedad': None},
    {'temperatura': 100,  'humedad': 102},
]

#limpieza de datos
lecturas_validas = [lecturas for lecturas in lecturas 
                    if lecturas['temperatura']
                    is not None and -20 <= lecturas['temperatura'] <= 60
                    and lecturas['humedad'] is not None and lecturas['humedad'] <= 100
                    ]

print("lecturas limpias: 5",  lecturas_validas)
print("-" * 30)#el "-" * 30 es para separ visualmente en consola


#clasificacion

niveles_criticidad =[
    "critico" if (lecturas['temperatura'] > 50 or lecturas['humedad'] > 90)
    else "normal" for lecturas in lecturas_validas
]

print("niveles de criticidaad:", niveles_criticidad)
```
