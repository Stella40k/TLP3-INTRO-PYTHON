# Operadores Lógicos en Python

Los operadores lógicos en Python se utilizan para combinar expresiones booleanas y evaluar condiciones.

## Operador `and`

Devuelve `True` si **ambas** expresiones son verdaderas.

| Expresión | Resultado |
|-----------|----------|
| True and True | True |
| True and False | False |
| False and True | False |
| False and False | False |

### Ejemplo:

```python
x = 5
resultado = (x > 0) and (x < 10)
print(resultado)  # True
```

````python
if True and True:
    print("True")
````

---

## Operador `or`

Devuelve `True` si **al menos una** de las expresiones es verdadera.

| Expresión | Resultado |
|-----------|----------|
| True or True | True |
| True or False | True |
| False or True | True |
| False or False | False |

### Ejemplo:

```python
edad = 25
resultado = (edad < 18) or (edad >= 65)
print(resultado)  # False
```

````python
if False or False:
    print("True")
````

---

## Operador `not`

Invierte el valor de la expresión: `True` se convierte en `False` y viceversa.

| Expresión | Resultado |
|-----------|----------|
| not True | False |
| not False | True |

### Ejemplo:

```python
es_estudiante = False
resultado = not es_estudiante
print(resultado)  # True
```

````python
if not True:
    print("True")
````
